Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
with open('popular-languages.csv', 'w', newline='') as csvfile:
    fieldnames = ['Language', 'Average Annual Salary']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for lang, salary in zip(languages, salaries):
        writer.writerow({'Language': lang, 'Average Annual Salary': salary})

print("Data has been successfully scraped and saved into 'popular-languages.csv'.")
```

```python
from bs4 import BeautifulSoup
import requests
import csv

# URL to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"

# Download the webpage at the url
data = requests.get(url).text

# Create a soup object
soup = BeautifulSoup(data, "html.parser")

# Scrape the Language name and annual average salary
languages = []
salaries = []

table = soup.find('table')  # Assuming the table contains the data
for row in table.find_all('tr')[1:]:  # Skip the header row
    cols = row.find_all('td')
    language = cols[1].text.strip()  # Assuming language name is in the second column
    salary = cols[3].text.strip().replace('$', '').replace(',', '')  # Remove '$' and ',' from the salary
    languages.append(language)
    salaries.append(int(salary))  # Convert salary to an integer

# Find the index of the language with the highest salary
index_highest_salary = salaries.index(max(salaries))

# Print the language with the highest average annual salary
print(f"The language with the highest average annual salary is: {languages[index_highest_salary]}")
```
