## Hands on Lab


Import pandas module.


```python
import pandas as pd
```

Load the dataset into a dataframe.


```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv")
```

## Finding duplicates


In this section you will identify duplicate values in the dataset.


 Find how many duplicate rows exist in the dataframe.


```python
# your code goes here
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
# Find the number of duplicate values in the column 'Respondent'
duplicate_respondents = df['Respondent'].duplicated().sum()
print(f"Number of duplicate values in 'Respondent': {duplicate_respondents}")
```

## Removing duplicates


Remove the duplicate rows from the dataframe.


```python
# your code goes here
df = df.drop_duplicates()
```

Verify if duplicates were actually dropped.


```python
# your code goes here
print("Number of duplicate rows after removal:", df.duplicated().sum())
```

```python
# Question 1 After removing the duplicate rows, how many rows are there in the dataset? 
rows_after_removing_duplicates = len(df)
print(f"Number of rows in the dataset after removing duplicates: {rows_after_removing_duplicates}")
```

```python
# Question 2 After removing the duplicate rows, how many unique rows are there in the column Respondent? 
unique_respondents = df['Respondent'].nunique()
print(f"Number of unique rows in the 'Respondent' column after removing duplicates: {unique_respondents}")
```

## Finding Missing values


Find the missing values for all columns.


```python
# your code goes here
missing_values_all_columns = df.isnull().sum()
print("Missing values for all columns:")
print(missing_values_all_columns)
```

Find out how many rows are missing in the column 'WorkLoc'


```python
# your code goes here
missing_values_workloc = df['WorkLoc'].isnull().sum()
print(f"Number of missing values in 'WorkLoc': {missing_values_workloc}")
```

```python
# Question 1 After removing the duplicate rows, how many blank rows are there under the column EdLevel?
blank_rows_edlevel = df['EdLevel'].isnull().sum()
print(f"Number of blank rows in the 'EdLevel' column after removing duplicates: {blank_rows_edlevel}")
```

```python
# Question 2 After removing the duplicate rows, how many rows are missing under the column Country?
missing_rows_country = df['Country'].isnull().sum()
print(f"Number of rows missing in the 'Country' column after removing duplicates: {missing_rows_country}")
```

## Imputing missing values


Find the  value counts for the column WorkLoc.


```python
# your code goes here

workloc_value_counts = df['WorkLoc'].value_counts()
```

Identify the value that is most frequent (majority) in the WorkLoc column.


```python
#make a note of the majority value here, for future reference
majority_value = workloc_value_counts.idxmax()
print(f"The majority value in 'WorkLoc' is: {majority_value}")
```

Impute (replace) all the empty rows in the column WorkLoc with the value that you have identified as majority.


```python
# your code goes here
df['WorkLoc'].fillna(majority_value, inplace=True)
```

```python
# What is the majority category under the column Employment?
majority_category_employment = df['Employment'].mode().iloc[0]
print(f"The majority category in the 'Employment' column is: {majority_category_employment}")
```

```python
# Under the column " UndergradMajor", which category has the minimum number of rows?
min_category_undergrad_major = df['UndergradMajor'].value_counts().idxmin()
min_category_count = df['UndergradMajor'].value_counts().min()

print(f"The category with the minimum number of rows in 'UndergradMajor' is: {min_category_undergrad_major}")
print(f"Number of rows for this category: {min_category_count}")
```

After imputation there should ideally not be any empty rows in the WorkLoc column.


Verify if imputing was successful.


```python
# your code goes here
print("Number of missing values in 'WorkLoc' after imputation:", df['WorkLoc'].isnull().sum())
```

## Normalizing data


There are two columns in the dataset that talk about compensation.

One is "CompFreq". This column shows how often a developer is paid (Yearly, Monthly, Weekly).

The other is "CompTotal". This column talks about how much the developer is paid per Year, Month, or Week depending upon his/her "CompFreq". 

This makes it difficult to compare the total compensation of the developers.

In this section you will create a new column called 'NormalizedAnnualCompensation' which contains the 'Annual Compensation' irrespective of the 'CompFreq'.

Once this column is ready, it makes comparison of salaries easy.


<hr>


List out the various categories in the column 'CompFreq'


```python
# your code goes here
compfreq_categories = df['CompFreq'].unique()
print("CompFreq categories:", compfreq_categories)
```

Create a new column named 'NormalizedAnnualCompensation'. Use the hint given below if needed.


Double click to see the **Hint**.

<!--

Use the below logic to arrive at the values for the column NormalizedAnnualCompensation.

If the CompFreq is Yearly then use the exising value in CompTotal
If the CompFreq is Monthly then multiply the value in CompTotal with 12 (months in an year)
If the CompFreq is Weekly then multiply the value in CompTotal with 52 (weeks in an year)

-->


```python
# your code goes here
df['NormalizedAnnualCompensation'] = df.apply(lambda row: row['CompTotal'] * 12 if row['CompFreq'] == 'Monthly' else row['CompTotal'] if row['CompFreq'] == 'Yearly' else row['CompTotal'] * 52, axis=1)
```

```python
# How many unique values are there in the CompFreq column?  
unique_values_compfreq = df['CompFreq'].nunique()
print(f"The number of unique values in the 'CompFreq' column is: {unique_values_compfreq}")
```

```python
# After removing the duplicate rows, how many respondents are being paid yearly?
yearly_paid_count = df[df['CompFreq'] == 'Yearly']['Respondent'].nunique()
print(f"The number of respondents being paid yearly after removing duplicates is: {yearly_paid_count}")
```

```python
# What is the median NormalizedAnnualCompensation? 
median_normalized_compensation = df['NormalizedAnnualCompensation'].median()
print(f"The median NormalizedAnnualCompensation is: {median_normalized_compensation}")
```
